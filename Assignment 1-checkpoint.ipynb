```python
# Generating Datasets

"""
Function to generate an m+1 dimensional data set, of size n, consisting of m continuous independent
variables (X) and one dependent variable (Y) 
"""

import numpy as np
import random

def generator(m,n,sigma):
    rg = np.random.default_rng(42)
    beta = rg.random((m+1,1))
    pre_x = np.random.rand(n,m)
    xzero = np.ones((n,1))
    x = np.concatenate((xzero,pre_x),axis=1)
    e = np.random.normal(0,sigma,size = (n,1))
    
    pre_y = np.dot(x,beta)
    y = pre_y + e
    return x,y,beta
```

```python
# Implementing Linear Regression
import math

def cost(y,y1):
    distance = y - y1
    sums = np.mean(distance**2)
    return np.sqrt(sums)

def improvement(y,ypred,x,beta):
    costvalue = cost(y,ypred)
    for i in range(x.shape[1]):
        betavalue = (beta[i])
        distance = y - ypred
        improvement = -np.dot(x.T,distance)/(costvalue*x.shape[0])
    return improvement

def linreg(x,y,k,tau,lam):
    rg = np.random.default_rng()
    beta = rg.random((x.shape[1],1))
    ypred = np.dot(x,beta)
    initcost = 0
    costvalue = cost(y,ypred)
    for i in range(k):
        if  abs(costvalue - initcost) > tau:
            initcost = costvalue
            impro = improvement(y,ypred,x,beta)
            beta = beta - lam*impro
            ypred = np.dot(x,beta)
            costvalue = cost(y,ypred)
        else:
            break
    return beta,costvalue
```

```python
import matplotlib.pyplot as plt
nvalues = np.arange(10,10000,100)
sigmavalues = np.arange(0,2,0.1)
costlist_n = []
costlist_sigma = []
for i in nvalues:
    x,y,beta = generator(10,int(i),0.1)
    beta_up,costvalue = linreg(x,y,10000,1e-6,0.01)
    costlist_n.append(costvalue)
for j in sigmavalues : 
    x,y,beta = generator(10,1000,j)
    beta_up,costvalue = linreg(x,y,10000,1e-6,0.01)
    costlist_sigma.append(costvalue)
    
plt.plot(nvalues,costlist_n,marker = ".")
plt.xlabel("n")
plt.ylabel("Cost Function")
plt.show()
plt.plot(sigmavalues,costlist_sigma, marker = ".")
plt.xlabel("sigma")
plt.ylabel("Cost Function")
plt.show()
```

- The final value of the cost function does not change significantly for increasing n.
- For increasing sigma, the value of cost function also increases. This is due to increasing inherent randomness of the sample data.
